#### Ejercicio 10. ¿Son los promedios de precio por metro cuadrado (price/m2) de "Valdemorillo" y "Galapagar" los mismos? (★★☆)

Imprime ambos promedios de precio y escribe una conclusion acerca de ellos.

Pista: Crea una nueva columna llamada `pps` (price per square) y luego analiza los valore.s

```python
valdemorillo_ds = ds[ds['level5'] == 'Valdemorillo'].copy()
galapagar_ds = ds[ds['level5'] == 'Galapagar'].copy()

valdemorillo_ds.loc[:, 'pps'] = valdemorillo_ds['price'] / valdemorillo_ds['surface']
galapagar_ds.loc[:, 'pps'] = galapagar_ds['price'] / galapagar_ds['surface']

pps_valdemorillo = valdemorillo_ds['pps'].mean()
pps_galapagar = galapagar_ds['pps'].mean()

print("El precio promedio por m2 de Valdemorillo es", pps_valdemorillo)
print("El precio promedio por m2 de Galapagar es", pps_galapagar)
print("Los precios promedio por metro cuadrado entre las dos ubicaciones tienen una diferencia de 289 teniendo con mayor precio Galapagar")
```

#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

For the normalization method you can use the one you consider to, there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot

Hint: You can help yourself reviewing the multihist demo of Matplotlib

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

for poblacion in poblaciones_cinturon_sur:
    precios_poblacion = cinturon_sur_ds.loc[cinturon_sur_ds['level5'] == poblacion, 'price'].values.reshape(-1, 1)
    precios_normalizados = scaler.fit_transform(precios_poblacion)
    cinturon_sur_ds.loc[cinturon_sur_ds['level5'] == poblacion, 'price_normalized'] = precios_normalizados

plt.figure(figsize=(10, 6))
for poblacion in poblaciones_cinturon_sur:
    precios_normalizados_poblacion = cinturon_sur_ds.loc[cinturon_sur_ds['level5'] == poblacion, 'price_normalized']
    plt.hist(precios_normalizados_poblacion, bins=20, alpha=0.5, label=poblacion)

plt.title('Histogramas de Precios Normalizados por Población')
plt.xlabel('Precio Normalizado')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid(True)
plt.show()
```

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph? You must use the subset obtained in the question 14 (★★☆) 
Hint: make a scatter plot of each population using subplots.

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

for i, poblacion in enumerate(poblaciones_cinturon_sur):
       poblacion_ds = cinturon_sur_ds[cinturon_sur_ds['level5'] == poblacion]
       fila = i // 2
       columna = i % 2
       axs[fila, columna].scatter(poblacion_ds['surface'], poblacion_ds['price'], alpha=0.5)
       axs[fila, columna].set_title(poblacion)
       axs[fila, columna].set_xlabel('Superficie (m2)')
       axs[fila, columna].set_ylabel('Precio')
       axs[fila, columna].grid(True)

plt.tight_layout()
plt.show()
```
